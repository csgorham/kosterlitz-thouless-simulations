```python
#----------------------------------------------------------------------
#  INITIALIZE SIMULATION PARAMETERS
#----------------------------------------------------------------------
N       = 2**2             # size of the lattice, N x N
eqSteps = 2**6 #2**7       # number of MC sweeps for equilibration
mcSteps = 2**6 #2**7       # number of MC sweeps for calculation


n1 = 1.0/(mcSteps*N*N*N*N)
n2  =1.0/(mcSteps*mcSteps*N*N*N*N)
outputfilename = './data/XYZW_unfrustrated_x'+ str(N)+ '+.csv'

nt      = 2**8 #2**8        # number of temperature points
### we anticipate the BKT transition to be 2.6, and get temperatures to sample around there
tm = 2.6;    T=np.random.normal(tm, .64, nt)
T  = T[(T>0.8) & (T<4.2)];    nt = np.size(T)

Energy       = np.zeros(nt);   Magnetization  = np.zeros(nt)
SpecificHeat = np.zeros(nt);   Susceptibility = np.zeros(nt)

helicity_wx      = np.zeros(nt);  helicity_wy      = np.zeros(nt);  helicity_wz      = np.zeros(nt); 
helicity_xy      = np.zeros(nt);  helicity_yz      = np.zeros(nt);  helicity_xz      = np.zeros(nt); 

#----------------------------------------------------------------------
#  MAIN PART OF THE CODE
#----------------------------------------------------------------------
config = Init_theta(N)
config1 = Init_theta1(N)
config2 = Init_theta2(N)


for m in range(len(T)):
    E1 = M1 = E2 = M2 = M1_1=M1_2=M1_3=0
    iT=1.0/T[m]; iT2=iT*iT;
    #e_wx1 = s_wx1 = e_wy1 = s_wy1 = e_wz1 = s_wz1 =  e_yz1 = s_yz1 =  e_xz1 = s_xz1 =  e_xy1 = s_xy1= 0

    
    '''equilibrate'''
    for i in range(eqSteps):        
        [config,config1,config2] = mcmove(config,config1,config2, iT)   # Monte Carlo moves

    '''simulate'''
    for i in range(mcSteps):
        [config,config1,config2] = mcmove(config,config1,config2, iT) # Monte Carlo moves        
        Ene = calcEnergy(config,config1,config2)                      # calculate the energy
        Mag = calcMag(config,config1,config2)                         # calculate the magnetisation
        
        
        #[e_wx, s_wx, e_wy, s_wy, e_wz, s_wz, e_xy, s_xy, e_xz, s_xz, e_yz, s_yz] = \
        #calcHelicity(config,config1,config2, N)

        E1 += Ene
        M1 += Mag
        M2 += Mag*Mag 
        E2 += Ene*Ene
        
       
        #e_wx1 += e_wx
        #s_wx1 += s_wx
        #e_wy1 += e_wy
        #s_wy1 += s_wy        
        #e_wz1 += e_wz
        #s_wz1 += s_wz
        #e_yz1 += e_yz
        #s_yz1 += s_yz
        #e_xz1 += e_xz
        #s_xz1 += s_xz
        #e_xy1 += e_xy
        #s_xy1 += s_xy
        
    E_mean = E1/mcSteps/(N**4)
    M_mean = M1/mcSteps/(N**4)
    Esq_mean = E2/mcSteps/(N**8)
    Msq_mean = M2/mcSteps/(N**8)
    Energy[m]         =  E_mean
    Magnetization[m]  =  M_mean
    SpecificHeat[m]   = (Esq_mean - E_mean**2)*iT2
    Susceptibility[m] = (Msq_mean - M_mean**2)*iT
        
    #helicity_wx[m] = n1*e_wx1 - n2*s_wx1*s_wx1*iT
    #helicity_wy[m] = n1*e_wy1 - n2*s_wy1*s_wy1*iT
    #helicity_wz[m] = n1*e_wz1 - n2*s_wz1*s_wz1*iT

        
#write data file for simulation
outputfile=open(outputfilename,"w")
data_file=csv.writer(outputfile, delimiter=",")
for m in range(len(T)):
    #data_file.writerow([ T[m], Energy[m], Magnetization[m], SpecificHeat[m], Susceptibility[m], helicity_wx[m], helicity_wy[m] , helicity_wz[m]  ])
    data_file.writerow([ T[m], Energy[m], Magnetization[m], SpecificHeat[m], Susceptibility[m]])


outputfile.close()
print('done')
```
