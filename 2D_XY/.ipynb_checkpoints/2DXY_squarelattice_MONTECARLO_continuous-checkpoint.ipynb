BKT transition temperature (unfrustrated) ~ 0.8 J/Kb...

BKT transition temperature (fully frustrated) ~ 0.6 J/Kb...

```python
%matplotlib inline
from __future__ import division
import csv
import numpy as np
from numpy import linalg as LA
import matplotlib.pyplot as plt
from numpy.random import rand

J = 1 # J>0 to make it ferromagnetic


'''Intitialize the XY network'''
def Init(N):
    return np.random.rand(N, N)*2*np.pi   ### range is -pi to pi

'''Monte Carlo move using Metropolis algorithm '''
def mcmove(config, beta):
    for i in np.arange(N**2):
        #for j in np.arange(N):
                a = np.random.randint(0, N)    ### x
                b = np.random.randint(0, N)    ### y 
                s =  config[a, b]

                if a%2 == 0:
                    #vertlink = np.pi   ### frustrated
                    vertlink = 0   ### unfrustrated
                else:
                    vertlink = 0       ### unfrustrated

                """Calculate energy over four nearest-neighbors on square lattice"""
                nb = -J*( (np.cos(s-config[(a+1)%N][b]))+\
                                 ( np.cos(s-config[(a-1)%N][b]))+\
                                 (np.cos(s-config[a][(b-1)%N] - vertlink)) + \
                                 (np.cos(s-config[a][(b+1)%N] - vertlink)) )


                flips = np.random.rand(1)*2*np.pi
                      
                nb2 = -J*(  (np.cos(flips-config[(a+1)%N][b])) +\
                                ( np.cos(flips-config[(a-1)%N][b])) +\
                                (np.cos(flips-config[a][(b-1)%N]- vertlink)) +\
                                (np.cos(flips-config[a][(b+1)%N] - vertlink)) )
                       
                       
                cost = nb2-nb
                
                if nb2 < nb: 
                    s = flips
                elif rand() < np.exp(-cost*beta):
                    s = flips
                config[a, b] = s
    return config

'''Energy of a given configuration'''
def calcEnergy(config):
    energy = 0;   
    
    for i in np.arange(N):
        for j in np.arange(N):
            S = config[i,j]
            
            if i%2 == 0:
                #IF FRUSTRATED: vertlink = np.pi   
                vertlink = 0   ### unfrustrated
            else:
                vertlink = 0       ### unfrustrated
            
            """Calculate energy over four nearest-neighbors on square lattice"""
            energy = energy - J*(  (np.cos(S-config[(i+1)%N, j])) + (np.cos(S-config[(i-1)%N, j])) + \
                + (np.cos(S-config[i,(j+1)%N]- vertlink)) + (np.cos(S-config[i,(j-1)%N]- vertlink))  )
                    

    return energy/2.

'''Magnetization of a given configuration'''
def calcMag(config):
    
    magx = np.sum(np.cos(config))
    magy = np.sum(np.sin(config))
    mag = np.array([magy,magx])
    
    mag1 = np.array([magx,magy])
    return   LA.norm(mag)

def calculate_e_helicity(config):
    """https://arxiv.org/pdf/1402.5017.pdf"""
    e=0
    for i in np.arange(N):
        for j in np.arange(N):
            e += (np.cos(config[i,j]- config[(i-1)%N,j]))
    return e

def calculate_s_helicity(config):
    """https://arxiv.org/pdf/1402.5017.pdf"""
    s=0
    for i in np.arange(N):
        for j in np.arange(N):
            s += (np.sin(config[i,j]- config[(i-1)%N,j]))
    return s
```

```python
#----------------------------------------------------------------------
#  INITIALIZE SIMULATION PARAMETERS
#----------------------------------------------------------------------
N       = 2**4     # size of the lattice, N x N
eqSteps = 2**8     # number of MC sweeps for equilibration
mcSteps = 2**8     # number of MC sweeps for calculation
outputfilename = './data/XY_unfrustrated_x'+ str(N) +'.csv'

size_xy = N**2 


nt      = 2**7     # number of temperature points
"""define temperature range"""
tm = 0.89213;     ### 0.89213 is the anticipated BKT transition temperature see: 
T=np.random.normal(tm, .65, nt)
T  = T[(T>0.4) & (T<1.7)];    
nt = np.size(T)

Energy       = np.zeros(nt);   Magnetization  = np.zeros(nt) ; 
SpecificHeat = np.zeros(nt);   Susceptibility = np.zeros(nt);  helicity = np.zeros(nt);

invtotalnum  = 1.0/size_xy 
n1, n2  = 1.0/(mcSteps)*invtotalnum, 1.0/(mcSteps*mcSteps)*invtotalnum

#----------------------------------------------------------------------
#  MAIN PART OF THE CODE
#----------------------------------------------------------------------
for m in range(len(T)):
    E1 = M1 = E2 = M2 = Y_e = Y_s = 0

    config = Init(N)
    iT=1.0/T[m]; iT2=iT*iT;

    for i in range(eqSteps):         # equilibrate
        mcmove(config, iT)           # Monte Carlo moves

    for i in range(mcSteps):
        mcmove(config, iT)           
        Ene = calcEnergy(config)     # calculate the energy
        Mag = calcMag(config)        # calculate the magnetisation
        hel_s = calculate_s_helicity(config)
        hel_e = calculate_e_helicity(config)
        
        E1 += Ene
        M1 += Mag
        M2 += Mag*Mag 
        E2 += Ene*Ene
        Y_e += hel_e
        Y_s += hel_s**2

        
    E_mean = E1/mcSteps/size_xy
    M_mean = M1/mcSteps/size_xy
    Esq_mean = E2/mcSteps/size_xy**2
    Msq_mean = M2/mcSteps/size_xy**2
    
    Y_e_mean = Y_e/mcSteps/size_xy
    Y_s_mean = Y_s/mcSteps/size_xy


    Energy[m]         =  E_mean 
    Magnetization[m]  =  M_mean 
    SpecificHeat[m]   = (Esq_mean - E_mean**2)*iT2
    Susceptibility[m] = (Msq_mean - M_mean**2)*iT
    helicity[m] = Y_e_mean - size_xy*iT*Y_s_mean


#write data file for simulation
outputfile=open(outputfilename,"w")
data_file=csv.writer(outputfile, delimiter=",")
for m in range(len(T)):
    data_file.writerow([ T[m], Energy[m], Magnetization[m], SpecificHeat[m], Susceptibility[m], helicity[m]])
outputfile.close()
    
print('done')
```
