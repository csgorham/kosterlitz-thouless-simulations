BKT transition temperature (fully frustrated) ~ 0.6 J/Kb...

```python
%matplotlib inline
from __future__ import division
import csv
import numpy as np
from numpy import linalg as LA
import matplotlib.pyplot as plt
from numpy.random import rand

##To simlify: J = k_B = 1 ----- J>0 to make it ferromagnetic


'''Intitialize the XY network'''
def Init(N):
    return np.random.rand(N, N)*2*np.pi   ### range is -pi to pi

'''Monte Carlo move using Metropolis algorithm '''
def mcmove(config, beta, L):
    for i in np.arange(L**2):
        #for j in np.arange(N):
                a = np.random.randint(0, L)    ### x
                b = np.random.randint(0, L)    ### y 
                s =  config[a, b]

                if a%2 == 0:
                    #vertlink = np.pi   ### frustrated
                    vertlink = 0   ### unfrustrated
                else:
                    vertlink = 0       ### unfrustrated

                """Calculate energy over four nearest-neighbors on square lattice"""
                nb = -( (np.cos(s-config[(a+1)%L][b]))+\
                                 ( np.cos(s-config[(a-1)%L][b]))+\
                                 (np.cos(s-config[a][(b-1)%L] - vertlink)) + \
                                 (np.cos(s-config[a][(b+1)%L] - vertlink)) )


                flips = np.random.rand(1)*2*np.pi
                      
                nb2 = -(  (np.cos(flips-config[(a+1)%L][b])) +\
                                ( np.cos(flips-config[(a-1)%L][b])) +\
                                (np.cos(flips-config[a][(b-1)%L]- vertlink)) +\
                                (np.cos(flips-config[a][(b+1)%L] - vertlink)) )
                       
                       
                cost = nb2-nb
                
                if nb2 < nb: 
                    s = flips
                elif rand() < np.exp(-cost*beta):
                    s = flips
                config[a, b] = s
    return config

'''Energy of a given configuration'''
def calcEnergy(config,L):
    energy = 0;   
    
    for i in np.arange(L):
        for j in np.arange(L):
            S = config[i,j]
            
            if i%2 == 0:
                #IF FRUSTRATED: vertlink = np.pi   
                vertlink = 0   ### unfrustrated
            else:
                vertlink = 0       ### unfrustrated
            
            """Calculate energy over four nearest-neighbors on square lattice"""
            energy = energy - ((np.cos(S-config[(i+1)%L, j])) + (np.cos(S-config[(i-1)%L, j])) + \
                + (np.cos(S-config[i,(j+1)%L]- vertlink)) + (np.cos(S-config[i,(j-1)%L]- vertlink))  )
                    

    return energy/2.

'''Magnetization of a given configuration'''
def calcMag(config):
    
    magx = np.sum(np.cos(config))
    magy = np.sum(np.sin(config))
    mag = np.array([magy,magx])
    
    return   LA.norm(mag)

def calculate_e_helicity(config,L):
    """https://arxiv.org/pdf/1402.5017.pdf"""
    e=0
    for i in np.arange(L):
        for j in np.arange(L):
            e += (np.cos(config[i,j]- config[(i-1)%L,j]))
    return e

def calculate_s_helicity(config,L):
    """https://arxiv.org/pdf/1402.5017.pdf"""
    s=0
    for i in np.arange(L):
        for j in np.arange(L):
            s += (np.sin(config[i,j]- config[(i-1)%L,j]))
    return s


def MONTE_CARLO_SIMULATION(L, min_T=0.1, max_T=2.5, step_T=25, eqSteps=5000, mcSteps=80000):
#----------------------------------------------------------------------
#  INITIALIZE SIMULATION PARAMETERS
#----------------------------------------------------------------------
    outputfilename = './data/XY_unfrustrated_x'+ str(L) +'.csv'
    size_xy = L**2
    
    Energy       = np.zeros(step_T);   Magnetization  = np.zeros(step_T) ; 
    SpecificHeat = np.zeros(step_T);   Susceptibility = np.zeros(step_T);  helicity = np.zeros(step_T);
    T_range = np.zeros(step_T); 

    
    by_T = (max_T - min_T)/step_T
    for step in range(step_T):
        T_range[step] = by_T + step* by_T
        

#    invtotalnum  = 1.0/size_xy 
#    n1, n2  = 1.0/(mcSteps)*invtotalnum, 1.0/(mcSteps*mcSteps)*invtotalnum

    
#----------------------------------------------------------------------
#  MAIN PART OF THE CODE
#----------------------------------------------------------------------
    for m in range(step_T):
        E1 = M1 = E2 = M2 = Y_e = Y_s = 0

        config = Init(L)
        iT=1.0/T_range[m]; iT2=iT*iT;

        for i in range(eqSteps):         # equilibrate
            mcmove(config, iT, L)           # Monte Carlo moves

        for i in range(mcSteps):
            mcmove(config, iT, L)           
            Ene = calcEnergy(config,L)     # calculate the energy
            Mag = calcMag(config)        # calculate the magnetisation
            hel_s = calculate_s_helicity(config,L)
            hel_e = calculate_e_helicity(config,L)
        
            E1 += Ene
            M1 += Mag
            M2 += Mag*Mag 
            E2 += Ene*Ene
            Y_e += hel_e
            Y_s += hel_s**2

        
        E_mean = E1/mcSteps/size_xy
        M_mean = M1/mcSteps/size_xy
        Esq_mean = E2/mcSteps/size_xy**2
        Msq_mean = M2/mcSteps/size_xy**2
    
        Y_e_mean = Y_e/mcSteps/size_xy
        Y_s_mean = Y_s/mcSteps/size_xy


        Energy[m]         =  E_mean 
        Magnetization[m]  =  M_mean 
        SpecificHeat[m]   = (Esq_mean - E_mean**2)*iT2
        Susceptibility[m] = (Msq_mean - M_mean**2)*iT
        helicity[m] = Y_e_mean - size_xy*iT*Y_s_mean


    #write data file for simulation
    outputfile=open(outputfilename,"w")
    data_file=csv.writer(outputfile, delimiter=",")
    for m in range(len(T)):
        data_file.writerow([ T[m], Energy[m], Magnetization[m], SpecificHeat[m], Susceptibility[m], helicity[m]])
    outputfile.close()
    
    print('finished for L={}'.format(L))
    return T_range, Energy, Magnetization, SpecificHeat, Susceptibility, helicity
```

<h3> SIMULATION

```python
#MONTE_CARLO_SIMULATION(L, min_T, max_T, step_T, eqstep, mcstep)
#returns T_range, E, M, M_sus, heatC, Y_helicity

T_range_10, E_10, M_10, heatC_10, M_sus_10, Y_helicity_10 = MONTE_CARLO_SIMULATION(L=10, step_T=32, eqSteps=80000)
T_range_30, E_30, M_30, heatC_30, M_sus_30, Y_helicity_30 = MONTE_CARLO_SIMULATION(L=30, step_T=32, eqSteps=80000)
T_range_60, E_60, M_60, heatC_60, M_sus_60, Y_helicity_60 = MONTE_CARLO_SIMULATION(L=60, step_T=32, eqSteps=80000)
T_range_100, E_100, M_100, heatC_100, M_sus_100, Y_helicity_100 = MONTE_CARLO_SIMULATION(L=100, step_T=32, eqSteps=80000)
```
